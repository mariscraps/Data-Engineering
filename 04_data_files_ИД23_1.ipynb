3. По данным из файла `addres-book-q.xml` сформировать список словарей с телефонами каждого из людей.

````python
from nltk import flatten

from bs4 import BeautifulSoup

with open("addres-book-q.xml") as f:
  ab = BeautifulSoup(f, "xml")

# получаем список объектов с тегом address
addresses = ab.findAll('address')

# получаем список объектов с тегом name
names = ab.findAll("name")

name_list = []

phones_dict = {}

for i in names:
  # добавляем только текст
  name_list.append(i.text)

count = - 1

for i in addresses:
  count += 1
  phones = i.findAll("phone")

  for j in phones:
    if name_list[count] not in phones_dict:
      phones_dict[name_list[count]] = j.text
    else:
      phones_dict[name_list[count]] = flatten(list([phones_dict[name_list[count]], j.text]))


phones_dict
````

````python
import json
with open("contributors_sample.json", mode="r", encoding="utf-8") as f:
  data = json.load(f)


def find_user(username):
  for i in data:
    if username == i["username"]:
      print(i)
      return ""

  raise ValueError(f'Пользователь с username: {username} не найден.')


try:
  username = input("Введите username: ")
  print(find_user(username))

except ValueError as e:
  print(e)
````

### pickle

2.1 На основе файла `contributors_sample.json` создайте словарь следующего вида:
```
{
    должность: [список username людей, занимавших эту должность]
}
```

````python
from nltk import flatten
import json
with open("contributors_sample.json", mode="r", encoding="utf-8") as f:
  data = json.load(f)

all_jobs = {}

for i in data:
  for j in i["jobs"]:
    if j not in all_jobs:
      all_jobs[j] = [i["username"]]
    else:
      all_jobs[j] = flatten(list([all_jobs[j], i["username"]]))


all_jobs
````

### XML

3.1 По данным файла `steps_sample.xml` сформируйте словарь с шагами по каждому рецепту вида `{id_рецепта: ["шаг1", "шаг2"]}`. Сохраните этот словарь в файл `steps_sample.json`

````python
from nltk import flatten
import json

from bs4 import BeautifulSoup
with open("steps_sample.xml") as f:
  # инициализируем объект типа xml
  ab = BeautifulSoup(f, "xml")

recipes_dict = {}

# получаем список объектов с тегом recipe
recipes = ab.findAll('recipe')

# получаем список объектов с тегом id
ids = ab.findAll("id")

id_list = []

for i in ids:
  id_list.append(i.text)

count = - 1

for i in recipes:
  count += 1
  steps = i.findAll("step")

  for j in steps:
    if id_list[count] not in recipes_dict:
      recipes_dict[id_list[count]] = j.text
    else:
      recipes_dict[id_list[count]] = flatten(list([recipes_dict[id_list[count]], j.text]))

# вывод первых четырех ключей и значений полученного словаря
print(list(recipes_dict.keys())[0], list(recipes_dict.values())[0])
print(list(recipes_dict.keys())[1], list(recipes_dict.values())[1])
print(list(recipes_dict.keys())[2], list(recipes_dict.values())[2])
print(list(recipes_dict.keys())[3], list(recipes_dict.values())[3])

#сохранение результата в файл json
with open("steps_sample.json", mode="w", encoding="utf-8") as f:
  json.dump(recipes_dict, f, indent = 2)

# открываем этот файл, проверяем, что все сохранилось
with open("steps_sample.json", mode="r", encoding="utf-8") as f:
  for line in f:
    pass
    #print(line, end="")
````

3.2 По данным файла `steps_sample.xml` сформируйте словарь следующего вида: `кол-во_шагов_в_рецепте: [список_id_рецептов]`

````python
from nltk import flatten
new_recipes_dict = {}

for key in recipes_dict:
  number_of_steps = len(recipes_dict[key])
  if number_of_steps in new_recipes_dict:
    new_recipes_dict[number_of_steps] = flatten(list([new_recipes_dict[number_of_steps], key]))
  else:
    new_recipes_dict[number_of_steps] = key


# вывод первых четырех ключей и значений полученного словаря
print(list(new_recipes_dict.keys())[0], list(new_recipes_dict.values())[0])
print(list(new_recipes_dict.keys())[1], list(new_recipes_dict.values())[1])
print(list(new_recipes_dict.keys())[2], list(new_recipes_dict.values())[2])
print(list(new_recipes_dict.keys())[3], list(new_recipes_dict.values())[3])
````

3.3 Получите список рецептов, в этапах выполнения которых есть информация о времени (часы или минуты). Для отбора подходящих рецептов обратите внимание на атрибуты соответствующих тэгов.

````python
# модуль .etree.ElementTree предоставляет функциональность для создания, редактирования и парсинга XML-файлов
import xml.etree.ElementTree as ET

minutes_or_hours_included = []

# получаем список объектов с тегом recipe
recipes = ab.findAll('recipe')

for i in recipes:
  steps = i.findAll("step")

  for j in steps:
    xmlstr = str(j)
    # ET.fromstring() -- создает объект Element из строки
    root = ET.fromstring(xmlstr)

    if 'has_minutes' in root.attrib:
      minutes_or_hours_included.append(i.find("id").text)
      break

    elif 'has_hours' in root.attrib:
      minutes_or_hours_included.append(i.find("id").text)
      break


print(minutes_or_hours_included[:10])
````

3.4 Загрузите данные из файла `recipes_sample.csv` (__ЛР2__) в таблицу `recipes`. Для строк, которые содержат пропуски в столбце `n_steps`, заполните этот столбец на основе файла  `steps_sample.xml`. Строки, в которых столбец `n_steps` заполнен, оставьте без изменений.

````python
import pandas as pd
from bs4 import BeautifulSoup
import json

with open('steps_sample.xml') as f:
    ab = BeautifulSoup(f, 'xml')

# собираем всю информацию по каждому из рецептов, берем из XML только текст
ids = [id.text for id in [ab.find_all('id')][0]]
recipes = [ab.find_all('steps')][0]
result = {}
steps = []
for recipe in recipes:
    steps.append([i.text for i in recipe.find_all('step')])
result = {ids[i]: steps[i] for i in range(len(ids))}
with open('steps_sample.json', 'w') as file:
    json.dump(result, file, indent=4)

# находим количество шагов каждого из рецептов
count_steps = {ids[i]: len(steps[i]) for i in range(len(ids))}
count_steps

recipes = pd.read_csv('recipes_sample.csv', delimiter=',', parse_dates=['submitted'])

# .iterrows(). -- метод, который возвращает итератор, который генерирует индекс
# каждой строки и данные этой строки в виде объекта Series
# index -- индекс строки в таблице
# row -- объект Series, который содержит данные строки таблицы
for index, row in recipes.iterrows():

    # pd.isnull() -- проверяет наличие NaN
    if pd.isnull(row['n_steps']):
        recipe_id = str(row['id'])
        if recipe_id in count_steps:
          # .at работает также, как и loc., но может принимать только одну
          # строку и один столбец в качестве входных аргументов
            recipes.at[index, 'n_steps'] = count_steps[recipe_id]
recipes
````

3.5 Проверьте, содержит ли столбец `n_steps` пропуски. Если нет, то преобразуйте его к целочисленному типу и сохраните результаты в файл `recipes_sample_with_filled_nsteps.csv`

````python
# isna() -- аналогична isnull() -- проверяет, является ли значение ложным,
# то есть NaN - пустым
count_Nan_steps = len(recipes[recipes['n_steps'].isna()])
if count_Nan_steps == 0:
  # .astype('int') -- устанавливаем значениям recipes['n_steps'] тип int
    recipes['n_steps'] = recipes['n_steps'].astype('int')
recipes.to_csv('recipes_sample_with_filled_nsteps.csv', sep=',')
````
